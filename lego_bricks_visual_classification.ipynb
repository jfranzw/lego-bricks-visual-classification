## Part 1 -  Dataset

### Part 1-a  - Load the dataset
The provided dataset is composed by pictures of Lego bricks, taken at different angles:<br>
As an example the following picture:<br>
<img src="dataset/train/2357 Brick corner 1x2x2/201706171206-0001.png"/> 
corresponds to the png file in "dataset/train/2357 Brick corner 1x2x2/201706171206-0001.png"
- "201706171206-0001.png" is the name of the file;
- "2357 Brick corner 1x2x2" is the label of the Lego brick type;
- "train" means that this is the portion of the dataset you must use for the training
Inside the folder <b>dataset</b> you can find another subfolder <b>validatetest</b>.
The content inside of <b>validatetest</b> must be used (after a proper split) for validation during training and for testing during the evaluation.

```python
 # CODE OF PART 1-a - "Load the Dataset"
import os
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
import tensorflow as tf
import logging
logger = tf.get_logger()
logger.setLevel(logging.ERROR)

SIZE = 128

def loadPictures(path):
  for _,dirs,_ in os.walk(path):
    X=[];
    y=[];
    for dir in dirs:
        for _,_,files in os.walk(path+dir+"/"):
            for file in files:
                if file.find(".png")<0:
                       continue     
                print("Loading:"+path+dir+"/"+file)
                img = cv2.imread(path+dir+"/"+file,cv2.IMREAD_UNCHANGED)
                img_resize = cv2.resize(img,(SIZE,SIZE),interpolation = cv2.INTER_AREA)
                gray = cv2.cvtColor(img_resize,cv2.COLOR_BGR2GRAY )
                X.append(gray)
                y.append(dir)
                
    return X,y


X_train,y_train=loadPictures('./dataset/train/')
X,y = loadPictures('./dataset/validationandtest/')
X_validate,X_test,y_validate,y_test = train_test_split(X, y, test_size=0.80)
```

The <b>loadPictures()</b> function is used to load all the pictures present
in the folder whose path is given by the parameter passed in input to the
function.<br>
The <b>walk()</b> function present in the <b>os</b> module allows to iterate, using
a top-down or bottom-up approach (in this case top-down by default),
on the generated directory tree and on the files inside
of all directories.<br>
Only files representing a <b>png</b> image are considered.<br>
Images are uploaded using the <b>cv2</b> library (openCV), a
library that was designed for computer vision problems.<br>
In particular with the <b>imread()</b> function an image is loaded.<br>
The <b>IMREAD_UNCHANGED</b> flag is used to load the image including the
alpha channel.<br>
    Resizing is then performed with the <b>resize()</b> function,
which resize the image using, as dimensions,
    the value of the <b>SIZE</b> variable and as interpolation the
flag <b>INTER_AREA</b>: this flag is a fast method that gets the average of several pixels
and it's good for shrinking an image.<br>
Finally to convert the image to grayscale, it is used
    the <b>cvtColor</b> function with the <b>COLOR_BGR2GRAY</b> flag.<br>
The function returns two lists:<br>
    list <b>X</b>: list of images<br>
    list <b>y</b>: list of folder names that basically represent
all the classes of the model we're going to build.<br>
    The <b>loadPictures()</b> function thus constructed is used to obtain
the list of images to be used for training and the list of 
target variables.<br>
    The same is done for the <b>validationandtest</b> directory:
    the ouputs in this case are split with the <b>train_test_split</b> function
    of <b>sklearn</b> so as to obtain  80% of test instances (for model evaluation)
and 20% instances for validation (to validate the model during
training)

### Part 1-b Dataset  preprocessing
Since you are using the dataset for a classification task using Keras API you will need to perform some dataset tranformations:<br>
1) tranform all the images into grayscale having a final shape of (dim,dim,1)  dim<200;<br>
<img src="info.png">
2) encode properly the target features for train, test and validation<br>

```python
 # CODE OF PART 1-b - "Dataset  preprocessing"
from keras.utils.np_utils import to_categorical
from sklearn.preprocessing import LabelEncoder
#reshaping
X_train = np.asarray(X_train)
X_test = np.asarray(X_test)
X_validate = np.asarray(X_validate)
y_train = np.asarray(y_train)
y_test  = np.asarray(y_test)
y_validate = np.asarray(y_validate)
X_train = X_train.astype('float32')/255
X_test = X_test.astype('float32')/255
X_validate = X_validate.astype('float32')/255

X_train = X_train.reshape(len(X_train),SIZE,SIZE,1)
X_test =  X_test.reshape(len(X_test),SIZE,SIZE,1)
X_validate = X_validate.reshape(len(X_validate),SIZE,SIZE,1)
encoder = LabelEncoder()
y_trainHot = to_categorical(encoder.fit_transform(y_train), 16)
y_testHot  = to_categorical(encoder.fit_transform(y_test), 16)
y_validateHot = to_categorical(encoder.fit_transform(y_validate), 16)
```

At this point all the images have already been transformed 
into grayscale and resized with a final shape of 
<b>(SIZE,SIZE)</b> <b>SIZE</b> < 200,
by the <b>loadPictures()</b> function used in the previous section.<br>
As regards to the second point of part 1.b, first of all,
all lists have been transformed into numpy arrays using the <b>asarray()</b>
function of <b>numpy</b> module and all images are scaled.<br>
Subsequently with the <b>reshape()</b> function a new shape 
was given to the arrays used for training, evaluation and validation.<br>
The arrays containing the target variables instead were first
encoded using <b>LabelEncoder</b>.<br>
<b>LabelEncoder</b> is an encoder that encode target labels 
with value between 0 and n-1, where n stands for number of classes.<br>
Then with <b>to_categorical()</b> function, that values were converted to binary 
class matrix (one hot encoding).<br>
<b>to_categorical</b> is a <b>keras</b> function,<b>keras</b> only supports one-hot-encoding for data 
that has already been integer-encoded.<br>
This is way target variables were first transformed using <b>LabelEncoder</b>.<br>
<b>to_categorical</b> makes also target variables 
suitable for use with <b>categorical_crossentropy</b> loss function.

## Part 2 - Training

## Part 2.a - create and train a  working CNN
use Keras: with Convolutional and Pooling operations and a dense sub network at the end, use properly the preprocessed portions of the dataset (train and validation) and apply scaling.

## Part 2.b -  save the learned model
save the model to a file

```python
# 2.a
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
import numpy as np

batch_size = 64
num_classes = 16
epochs = 40
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3),
                 activation='relu',
                 input_shape=(SIZE,SIZE,1)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='softmax'))

model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adadelta(),
              metrics=['acc'])

model.fit(X_train, y_trainHot,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(X_validate, y_validateHot))
```

```python
#2.b  
model.save('model.h5')
```

## 2.a  comments##
In this part the build,the compilation and the training of the model is performed.<br>
### Model building
The type of model used is <b>Sequential</b>, which allows you to build the model layer by layer.<br>
The <b>add()</b> function allows you to add a new layer to the model.<br>
The first two layers called <b>Conv2D</b> are convolution layers that have to do with the input images.<br>
The first has 32 nodes, the second 64,both use a 3x3 filter matrix and the <b>Rectified Linear Activation</b> as activation function.<br>
The <b>input shape</b> flag of the first layer,indicates the shape of the input images <b>(SIZE, SIZE, 1)</b> where 1 indicates that the images are grayscale.<br>
<b>MaxPooling2D</b> is the layer whose task is to carry out a sample-based discretization process.<br>
The goal of this layer is to down-sample an input reducing its dimensionality.<br>
<b>pool_size</b> flag represents the size of the max pooling windows used to perform this operations.<br>
<b>Dropout</b> layer helps prevent overfitting setting (randomly) a fraction rate of input units to 0 at each update during training.<br> 
The <b>Flatten</b> layer is mainly used to connect convolutional layers and dense layers.<br>
<b>Dense</b> layer is the layer used for the model's output, in fact(as regards the last layer) the number of nodes that make it up, corresponds to all possible outcomes. Its activation function is the <b>softmax</b>, which allows us to provide an ouput that can be interpreted as a probability.
<br>
### Model compiling
After building the model layer by layer, it's time to compile it.<br>
With the <b>compile()</b> function, the model is compiled. Compilation requires three parameters: <b>loss</b>, <b>optimizer</b> and <b>metrics</b>.<br>
<b>loss</b> represents the cost function used which in this case is <b>categorical_crossentropy</b>.
The <b>optimizer</b> is used to check the learning rate of the model, which determines how fast the optimal weights are calculated.<br>
The <b>accuracy</b> is used as a metric to display the score of the model on the validation set during the training.<br>

### Model training
Model training is performed with the <b>fit()</b> function.<br>
This functions receives in input training data, target data and validation data (carefully preprocessed).<br>
The <b>batch size</b> is a <b>hyperparameter</b> that defines the number of samples to work through before updating the internal model parameters.<br>
    The number of <b>epochs</b> is a <b>hyperparameter</b> that defines the number times that the learning algorithm will work through the entire training dataset.<br>
    <b>validation_data</b> are data on which to evaluate the loss and any model metrics at the end of each epoch.<br>
    After playing with the <b>hyperparameters</b>, setting a relatively small <b>batch size</b> (64) and training the CNN for about 40 epochs, it was possible to obtain an accuracy on the set used for the validation of 1.0.

# Part 4 - Feature importance
## Part 4.a perform a pixel based feature importance analysis on the validation portion of the dataset
## Part 4.b plot the corresponding heatmap of pixel importance



```python
# code for part 4
#4.a
print(__doc__)

from time import time
import matplotlib.pyplot as plt

from sklearn.datasets import fetch_olivetti_faces
from sklearn.ensemble import ExtraTreesClassifier

# number of cores used to perform parallel fitting of classifier
n_jobs = 4

X, y = X_validate,y_validate
X = X.reshape(len(X),SIZE*SIZE)

# Build a forest and compute the pixel importances
t0 = time()
forest = ExtraTreesClassifier(n_estimators=1000,
                              max_features=128,
                              n_jobs=n_jobs,
                              random_state=0)

forest.fit(X, y)
print("done in %0.3fs" % (time() - t0))

#4.b
importances = forest.feature_importances_
importances = importances.reshape((SIZE, SIZE))

# Plot pixel importances
plt.matshow(importances, cmap=plt.cm.hot)
plt.title("Pixel importances with forests of trees")
plt.show()
```

## 4.a comments
In this section a pixel based feature importance analysis on the validation portion of the dataset is perfomed using a <b>parallel forest of trees</b>.<br>
Feature importance provides a score for each feature of the dataset.
The higher the score, the greater the importance of the feature for model output.<br>
A parallel forest of trees allow  to measure the relative importance of each feature.</br>
